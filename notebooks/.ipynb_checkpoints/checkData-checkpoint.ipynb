```python
import json
```

```python
data_file="../data/lotr_nocomment.json"
```

```python
with open(data_file, "r") as f:
    data = json.load(f)
```

```python
dates = data['dates']
```

```python
dates
```

```python
character_places = []
for date, _data in dates.items():
    obj ={k:v for k,v in _data['characters'].items()}
    obj["__date"] = date
    character_places.append(obj)
```
